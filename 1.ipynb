```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('class_demo_dataset.csv')
df.head()
```

```python
plt.scatter(df['study_hours'], df['quiz_score'],color='tomato')
plt.title('Study Hours vs Quiz Score')
plt.xlabel('Study Hours (per week)')
plt.ylabel('Quiz Score (%)')
#plt.grid(True)
plt.show()
```

Accuracy Loss curve

```python
df1 = pd.read_csv('training_metrics.csv')
df1
df1.plot(x='epoch', y=['train_loss', 'val_loss'])
plt.title('Loss')
plt.xlabel('Epoch');
plt.ylabel('Loss');
#plt.grid(True);
plt.show()
```

Accuracy

```python
df1.plot(x='epoch', y='accuracy')
plt.title('Accuracy')
plt.xlabel('Epoch');
plt.ylabel('Accuracy');
#plt.grid(True);
plt.show()
```
